**Food Delivery Sales & Customer Insights**

```python
import pandas as pd
import matplotlib.pyplot as plt

#Sample Food Delivery Dataset
data={
    'Order_ID':[101,102,103,104,105,106,107,108],
    'Time':['12:30','13:15','18:45','19:00','20:30','12:15','18:30','21:00'],
    'Location':['Chennai','Coimbatore','Chennai','Madurai','Chennai','Madurai','Coimbatore','Chennai'],
    'Item':['Pizza','Burger','Pizza','Biryani','Burger','Pizza','Biryani','Pizza'],
    'Amount':[250,180,300,220,200,280,240,350]
}
df=pd.DataFrame(data)

#Convert time to hour
df['Hour']=pd.to_datetime(df['Time']).dt.hour

#Peak Order Times
peak_hours=df['Hour'].value_counts().sort_index()

#Most Popular Dishes
popular_items=df['Item'].value_counts()

print("Peak Order Hours:")
print(peak_hours)

print("\nMost Popular Dishes:")
print(popular_items)

#Dashboard
plt.figure(figsize=(10,5))

#Peak Hours Chart
plt.subplot(1,2,1)
peak_hours.plot(kind='bar')
plt.title('Orders by Hour')
plt.xlabel('Hour')
plt.ylabel('Number of Orders')

#Popular Dishes Chart
plt.subplot(1,2,2)
popular_items.plot(kind='bar')
plt.title('Popular Food Items')
plt.xlabel('Item')
plt.ylabel('Orders')

plt.tight_layout()
plt.show()
```
