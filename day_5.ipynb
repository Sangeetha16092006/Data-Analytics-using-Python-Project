PRODUCT SALES PERFORMANCE

```python
from typing import ByteString
import numpy as np
products=["Laptop","Mobile","Tablet","Headphones","Smartwatch"]

#Sales units
sales=np.array([120,250,80,150,100])

#Total sales
print("TotalSales",np.sum(sales))

#Average sales
print("AverageSales",np.mean(sales))

#Highest sales
print("HighestSales",np.mean(sales))

#Lowest sales
print("LowestSales",np.min(sales))

#Best-selling product
best_index=np.argmax(sales)
print("Best Selling Product:",products[best_index])

#Product-wise sales
print("\nProduct Sales Report")
for i in range(len(products)):
  print(products[i],":",sales[i])
```
