**Handling Duplicates:**

```python
import pandas as pd
data={
    'Name':["John","Mary","John"],
    'Sales':[1000,1500,1000]
}
df=pd.DataFrame(data)
print(df.duplicated())
```

```python
import pandas as pd
data={
    'Name':["John","Mary","John"],
    'Sales':[1000,1500,1000]
}
df=pd.DataFrame(data)
df=df.drop_duplicates()
print(df)
```

**Rename Columns:**

```python
import pandas as pd
data={
    'emp_nm':["John","Mary"],
    'emp_sal':[25000,30000]
}
df=pd.DataFrame(data)
df.rename(columns={
    'emp_nm':'Employee Name',
    'emp_sal':'Employee Salary'
},inplace=True)
print(df)
```

**Data Filtering & Transformation:**

```python
import pandas as pd
data={
    "Region":["South","North","East","West","South","North"],
    "Product":["Laptop","Mobile","Laptop","Tablet","Mobile","Tablet"],
    "Sales":[50000,30000,45000,25000,35000,40000]
}
df=pd.DataFrame(data)
print(df)
```

```python
south_sales=df[df["Region"]=="south"]
print(south_sales)
```

```python
high_sales=df[df["Sales"]>30000]
print(high_sales)
```

**SORTING:**

```python
sorted_df=df.sort_values("Sales")
print(sorted_df)
```

**clean a messy sales dataset:**

```python
import pandas as pd
data={
    "Employee_ID":[101,102,103,104,105,106],
    "Name":["Keerthi","Sangeetha","Meenal","Divya","Priya","None"],
    "Age":[25,20,26,27,28,None],
    "Salary":[50000,20000,25000,12000,30000,None],
}
df=pd.DataFrame(data)
print(df)
print(df.isnull())
df=(df.drop_duplicates())
print(df)
df["Age"]=df["Age"].fillna(df["Age"].mean())
print(df)
df=df.dropna(subset=["Salary"])
print(df)
df=df.rename(columns={"Name":"Employee_Name","Age":"Employee_Age"})
print(df)
```

```python
region_sales=df.groupby("Region")["Sales"].sum()
print(region_sales)
```

```python
product_sales=df.groupby("Product")["Sales"].sum()
print(product_sales)
```
