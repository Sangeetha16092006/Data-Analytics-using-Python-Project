```python
def calculate_average(a,b,c):
  return(a+b+c)/3

try:
  a=int(input("Enter 1st number:"))
  b=int(input("Enter 2nd number:"))
  c=int(input("Enter 3rd number:"))
  print("calculate_average=",calculate_average(a,b,c))
except ValueError:
  print("Invalid Input")
```

```python
def attendance_percentage(attended,total):
  percentage=attended/total*100
  return round(percentage,2)

def check_eligibility(percent):
  if percent>=75:
    return "ELIGIBLE"
  else:
    return "NOT ELIGIBLE"

try:
  total=int(input("Total Classes:"))
  attended=int(input("Classes Attended:"))
  pct=attendance_percentage(attended,total)
  status=check_eligibility(pct)

  print(f"Attendance:{pct}%")
  print(f"status:{status}")

except valueError:
  print("Error:Enter valid numbers")
```
