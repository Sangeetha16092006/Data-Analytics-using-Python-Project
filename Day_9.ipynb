**pie chart:**

```python
import matplotlib.pyplot as plt
expenses=[5000,3000,2000]
categories=["Food","Travel","Shopping"]
plt.pie(expenses,
        labels=categories)
plt.show()
```

**Monthly sales dashboard:**

```python
import matplotlib.pyplot as plt

months=["Jan","Feb","Mar","Apr","May","Jun"]

sales=[10000,12000,15000,13000,18000,20000]

plt.figure(figsize=(18,5))# Adjusted figsize to accomodate three plots

#Line chart
plt.subplot(1,3,1)# Charged to 1 row, 3 columns,position 1
plt.plot(months,
         sales,marker='o')
plt.title("Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")

#Bar chart
plt.subplot(1,3,2)#Charged to 1row, 3 columns,position 2

plt.bar(months,
        sales,
        color="orange")

plt.title("Sales comparision")

plt.xlabel("Months")

plt.ylabel("Sales")

#Pie chart
plt.subplot(1,3,3)#Charged to 2 row, 3 columns,position 3

expenses=[5000,3000,2000]

categories=["Food","Travel","Shopping"]

plt.pie(expenses,
        labels=categories,
        autopct="%1.1f%%",colors=["pink","blue","green"])

plt.title("Expense Distribution")

plt.tight_layout()

plt.show()
```

**Histrogram:**

```python
import matplotlib.pyplot as plt
marks=[45,50,55,60,65,70,72,75,80,85,90,95]
plt.hist(marks,bins=5)
plt.title("Student Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of students")
plt.show()
```

**Scatter Plots:**

```python
import matplotlib.pyplot as plt
hours=[1,2,3,4,5,6,7,8]
marks=[35,40,50,60,65,75,85,90]
plt.scatter(hours,marks)
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

**Subplot:**

```python
import matplotlib.pyplot as plt
plt.subplot(1,2,1)
plt.plot([1,2,3],[10,20,30])
plt.title("Graph 1")
plt.subplot(1,2,2)
plt.plot([1,2,3],[30,20,18])
plt.title("Graph 2")
plt.show()
```
