**Handling Duplicates**

```python
import pandas as pd
data={
    'Name':["John","Mary","John"],
    'Sales':[1000,1500,1000]
}
df=pd.DataFrame(data)
print(df.duplicated())
```

```python
import pandas as pd
data={
    'Name':["John","Mary","John"],
    'Sales':[1000,1500,1000]
}
df=pd.DataFrame(data)
df=df.drop_duplicates()
print(df)
```

**Rename Columns:**

```python
import pandas as pd
data={
    'emp_nm':["John","Mary"],
    'emp_sal':[25000,30000]
}
df=pd.DataFrame(data)
df.rename(columns={
    'emp_nm':'Employee Name',
    'emp_sal':'Employee Salary',
},inplace=True)
print(df)
```

**Data Filtering & Transformation:**

```python
import pandas as pd
data={
    'Region':["South","North","East","West","South","North"],
    'Product':["Laptop","Mobile","Laptop","Tablet","Mobile","Tablet"],
    'Sales':[50000,30000,45000,25000,35000,40000]
}
df=pd.DataFrame(data)
print(df)
```

```python
South_sales=df[df["Region"]=="south"]
print(south_sales)
```
